```python
import pandas as pd
import matplotlib.pyplot as plt


# Define a function to create a bar plot for categorical features
def plot_bar(feature):
    counts = eda_data_df[feature].value_counts()
    plt.bar(counts.index, counts.values)
    plt.xticks(rotation=90)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.show()

# Define a function to create a histogram for numerical features
def plot_hist(feature):
    plt.hist(eda_data_df[feature], bins=20)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.show()
```
